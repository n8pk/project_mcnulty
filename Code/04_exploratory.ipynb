```python
import plotly
import pandas as pd
import numpy as np
import itertools

from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import confusion_matrix

import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
df_12 = pd.read_csv('/home/nate/ds/metis/class_work/projects/clean_data/df_12_altered.csv')
df_13 = pd.read_csv('/home/nate/ds/metis/class_work/projects/clean_data/df_13_altered.csv')
df_14 = pd.read_csv('/home/nate/ds/metis/class_work/projects/clean_data/df_14_altered.csv')
df_15 = pd.read_csv('/home/nate/ds/metis/class_work/projects/clean_data/df_15_altered.csv')
df_16 = pd.read_csv('/home/nate/ds/metis/class_work/projects/clean_data/df_16_altered.csv')
```

# Let's try to predict 2018

## make df for 18

For diversity and college, use 4-year average since that's the most extensive data we have

Years with medical is easier, just add 2 to anything with, and throw in the new ones from 2016 and 2017

```python
df_16['white-4'] = df_12['white_ratio']
df_16['college-4'] = df_12['college_ratio']


white_4_deltas = []
college_4_deltas = []

for idx, row in df_16.iterrows():
    
    white_4_delta = (row[2]-row[11])/row[11]
    if white_4_delta < 0:
        white_4_deltas.append((-(abs(white_4_delta-1)**0.25)))
    else:
        white_4_deltas.append(((white_4_delta+1)**0.25))
    
    college_4_delta = (row[3]-row[12])/row[12]
    college_4_deltas.append(((college_4_delta+1)**0.25))
    

df_16['white_4_deltas'] = white_4_deltas
df_16['college_4_deltas'] = college_4_deltas
```

```python
# 13, 14

white_18 = []
college_18 = []

for idx, row in df_16.iterrows():
    
    if row[13] < 0:
        white_18.append(row[2]/(abs(row[13]))**2)
    else:
        white_18.append(row[2]*row[13]**2)
        
    college_18.append(row[3]*row[14]**2)
```

```python
years_med_16 = np.array([ 0, 18,  6,  0, 20, 16,  4,  5,  0,  0, 16,  0,  3,  0,  0,  0,  0,\
                          0, 17,  2,  4,  8,  2,  0,  0, 12,  0, 16,  3,  6,  9,  2,  0,  0,\
                          0,  0, 18,  0, 10,  0,  0,  0,  0,  0, 12,  0, 18,  0,  0,  0])
```

```python
years_med_18 = []

for i in years_med_16:
    if i > 0:
        i+=2
        years_med_18.append(i)
    else:
        years_med_18.append(i)
```

```python
med16 = [3, 8, 33, 34, 37]
med17 = [47]
med18 = []

for idx, years in enumerate(years_med_18):
    if idx in med16:
        years+=2
    elif idx in med17:
        years+=1
    med18.append(years)
```

```python
rec_legal = df_16['rec_legal'].copy()
rec_legal[44] = float(rec_legal[44] + 1)
```

```python
geo = df_16['geography']

df_18 = pd.DataFrame(geo)
df_18['white_ratio'] = white_18
df_18['college_ratio'] = college_18
df_18['years_med'] = med18
df_18['rec_legal'] = rec_legal
```

## throw it in the models and viz

```python
test = df_18

train_years = [df_12, df_13, df_14, df_15, df_16]
train_whole = pd.concat(train_years).reset_index(drop=True)

X_train = np.array(train_whole[
    ['white_ratio', 'college_ratio', 'years_med']]).reshape(-1,3)
y_train = np.array(train_whole['rec_legal'])

X_test = np.array(test[
    ['white_ratio', 'college_ratio', 'years_med']]).reshape(-1,3)
y_test = np.array(test['rec_legal'])
```

```python
lr = LogisticRegression(penalty='l1', C=0.1, random_state=42)
lr.fit(X_train, y_train)
lr_pred = lr.predict_proba(X_test)

lr_probs = lr_pred[:, 1]
lr_names = df_16['geography']

df_lr_probs = pd.DataFrame(lr_names)
df_lr_probs['lr_probability'] = lr_probs
df_lr_probs = df_lr_probs.sort_values(by=['lr_probability'], ascending=False).reset_index()

#df_lr_probs
```

```python
gbc = GradientBoostingClassifier(n_estimators=6, learning_rate=0.1, max_features=None, max_depth= 2, random_state=42)
gbc.fit(X_train, y_train)
gbc_pred = gbc.predict_proba(X_test)

gbc_probs = gbc_pred[:, 1]
gbc_names = df_16['geography']

df_gbc_probs = pd.DataFrame(gbc_names)
df_gbc_probs['gbc_probability'] = gbc_probs
df_gbc_probs = df_gbc_probs.sort_values(by=['gbc_probability'], ascending=False).reset_index()

#df_gbc_probs
```

```python
plt_data = df_18.copy()

trues = plt_data.groupby(['rec_legal']).get_group(1)
falses = plt_data.groupby(['rec_legal']).get_group(0)

illegals = falses.groupby(['years_med']).get_group(0)


void = list(illegals['geography'])

medical = []
for idx, row in falses.iterrows():
    if row[0] not in void:
        medical.append(row)

medicals = pd.DataFrame(medical)
```

```python
preds = df_gbc_probs.head(12)
preds = preds['geography']

pred_points = []

for idx, row in plt_data.iterrows():
    if row[0] in list(preds):
        pred_points.append(row)
        
pred_df = pd.DataFrame(pred_points)
```

```python
plt.figure(figsize=(9,6))

plt.title('Wild Speculation ("Predicting" 2018)', fontweight='bold', fontsize=24)
plt.xlabel('% Population > 25 with a College Degree', fontweight='bold', fontsize=16)
plt.ylabel('Years with Medical Marijuana', fontweight='bold', fontsize=16)

plt.text(0.36, 15, 'Vermont', fontweight='bold', fontsize=12)
plt.text(0.435, 4.2, 'Massachusetts', fontweight='bold', fontsize=12)
plt.text(0.32, 16.3, 'Hawaii', fontweight='bold', fontsize=12)

plt.grid()
plt.xlim(0.2,0.5)
plt.ylim(-2,24)
plt.plot(trues['college_ratio'], trues['years_med'], 'go', label='Recreational', ms=10)
plt.plot(medicals['college_ratio'], medicals['years_med'], 'mo', label='Medical', ms=10)
plt.plot(illegals['college_ratio'], illegals['years_med'], 'ro', label='Illegal', ms=10)
plt.plot(pred_df['college_ratio'], pred_df['years_med'], 'h', markerfacecolor='None',
         markeredgecolor='k', markeredgewidth=2.5, ms=20, label='Predicted')
plt.legend(loc=(0.75,0.4), prop={'size': 12}, edgecolor='k');
```
