```python
def legal(df, states):
    
    '''
    make col for legal based on list of states
    '''
    
    legal = np.zeros((53,1))
    df['legal'] = legal

    for index, row in df.iterrows():
        if df.iloc[index][0] in states:
            df.at[index, 'legal'] = 1.0
            
    return df
```
