```python
def legal(df, states):
    
    '''
    make col for legal based on list of states
    '''
    
    df = df.drop(52)
    legal = np.zeros((52,1))
    df['legal'] = legal

    for index, row in df.iterrows():
        if df.iloc[index][0] in states:
            df.at[index, 'legal'] = 1.0
            
    return df
```

```python
df_12 = legal(df_12, med_12)
df_13 = legal(df_13, med_13)
df_14 = legal(df_14, med_14)
df_15 = legal(df_15, med_15)
df_16 = legal(df_16, med_16)
```
